## Lets make a perceptron, then build from there

```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
# Red is above the target function and blue is below

def blue(x,y):
    plot.plot(x, y, 'bo')

def red(x,y):
    plot.plot(x, y, 'ro')
```

Setting things up

```python
size = 20

# number
n = 100

xspace = np.linspace(0, size);

# Target Function
def TargetF(x):
    return 2*x+1

plot.plot(xspace, TargetF(xspace),'black')
```

```python
trainingset = nr.rand(n, n)*size*2
xs = trainingset[:,1]
ys = trainingset[:,0]
plot.plot(xs, ys,'bo')
```

```python
for i in xrange(n):
    y = ys[i]
    x = xs[i]
    if TargetF(x) > y:
        red(x, y)
    else:
        blue(x,y)

plot.plot(xspace, TargetF(xspace),'black')
```
